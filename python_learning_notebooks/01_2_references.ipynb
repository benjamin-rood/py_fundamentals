# Lesson 1.2: Variables Are References, Not Containers

**Duration:** ~1 hour

This is the most important lesson in the course. Understanding this one idea — that variables *refer to* objects rather than *containing* them — prevents a whole category of bugs that affect both human-written and AI-generated code. Here's a concrete example: an AI coding assistant writes `backup = patient_list` thinking it created a safety copy. But because variables are references, both names point to the same list — and modifying one corrupts the other. By the end of this lesson, you'll spot that mistake instantly.

## Assignment creates a reference

When you write `y = x`, you are **not** copying the value. You're making `y` refer to the **same object** that `x` refers to.

The square brackets `[1, 2, 3]` create a **list** — an ordered collection of values. We'll cover lists fully in Week 4. The key property here is that lists are **mutable** (the contents can be changed after creation).

````python
x = [1, 2, 3]
y = x
print(f"id(x) = {id(x)}")
print(f"id(y) = {id(y)}")
print(f"x is y: {x is y}")
````

`x is y` is `True`. There's one list in memory, and both `x` and `y` point to it.

This is called **aliasing** — two names for the same object.

## Identity vs. Equality

Python has two different checks:
- `is` — are these the **same object**? (same `id()`)
- `==` — do these have the **same value**?

````python
x = [1, 2, 3]
y = x            # Same object
z = [1, 2, 3]    # New object, same values

print(f"x == y: {x == y}")   # True — same value
print(f"x is y: {x is y}")   # True — same object
print()
print(f"x == z: {x == z}")   # True — same value
print(f"x is z: {x is z}")   # False — different object
print()
print(f"id(x) = {id(x)}")
print(f"id(y) = {id(y)}")
print(f"id(z) = {id(z)}")
````

`x` and `y` are the same object — same value AND same identity.

`x` and `z` have the same value but are different objects — they live at different memory addresses.

This distinction matters. We'll see exactly why below.

**The correct practice in real code:** use `==` to compare values (almost always what you want). Use `is` only to check for `None` — e.g. `if result is None:`. You'll rarely need `is` for anything else.

## Reassignment changes the reference

When you reassign a variable, you're making it point to a different object. The original object is unchanged.

````python
a = 10
b = a
print(f"Before: a = {a}, b = {b}")
print(f"Before: id(a) = {id(a)}, id(b) = {id(b)}")
print()

a = a + 1   # a now refers to a NEW object (11)
print(f"After:  a = {a}, b = {b}")
print(f"After:  id(a) = {id(a)}, id(b) = {id(b)}")
````

After `a = a + 1`:
- `a` now refers to a **new** integer object (`11`) — notice the different `id()`
- `b` still refers to the **original** object (`10`) — its `id()` hasn't changed

Reassignment moved the `a` label to a different object. It did not modify the `10` object. It did not affect `b`.

## Integers and strings are immutable

The `10` object cannot be changed after it's created. You can't turn a `10` into an `11` — you can only create a new `11` and point your variable at it.

Objects that **cannot be changed** are called **immutable.** Integers, floats, strings, and booleans are all immutable.

This is why `a = a + 1` creates a new object: you can't modify the existing `10`, so Python creates `11` and rebinds `a`.

This is a deliberate design choice. Because immutable objects can never change, Python can safely let multiple variables point to the same object (as you saw with small integer caching in Lesson 1.1). If integers *could* be changed, modifying one variable's value would silently change another's — producing bugs that are difficult to detect.

````python
# Strings are immutable too
text = "hello"
print(f"Before: text = {text}, id = {id(text)}")

text = text + " world"
print(f"After:  text = {text}, id = {id(text)}")
print("Different id — new object!")
````

## Why this matters for AI-generated code

When an AI coding agent writes code that passes data between functions, you need to know:
- Did the function get the **same object** or a **copy**?
- If it modifies the object, will the caller see the change?
- If it reassigns the parameter, does that affect anything outside?

## What you learned

This mental model — variables as labels pointing to objects — will explain a wide range of Python behaviours you'll encounter from here on. It's the single most useful concept in this course.

- Variables **refer to** objects — they don't contain them
- `y = x` makes `y` refer to the **same object** as `x` (aliasing)
- `is` checks identity (same object); `==` checks equality (same value)
- Reassignment makes a variable point to a different object
- Integers and strings are **immutable** — they can't be changed, only replaced
