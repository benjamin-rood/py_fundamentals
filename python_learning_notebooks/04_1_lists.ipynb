Lists are Python's primary mutable sequence type — ordered collections that can be modified after creation. AI-generated code uses them constantly. This lesson covers creation, access, mutation, and the aliasing trap.

## Creating and accessing lists

````python
numbers = [1, 2, 3, 4, 5]
print(numbers)
print(type(numbers))
````

````python
# Access by index — zero-indexed!
print(numbers[0])    # First element
print(numbers[2])    # Third element
print(numbers[-1])   # Last element
print(numbers[-2])   # Second to last
````

A common source of confusion: `numbers[2]` gives you the **third** element, not the second. The index counts how many positions from the start — position 0 is the first item. This convention comes from how computers store data: the index is the *offset* (distance) from the beginning, so the first element is 0 steps away.

Negative indices count backwards from the end: `-1` is the last element, `-2` is second-to-last, and so on. Internally, `numbers[-1]` is equivalent to `numbers[len(numbers) - 1]` — Python just does the subtraction for you.

````python
# Slicing — creates a new list
print(numbers[1:3])   # Elements 1 and 2 (stop is excluded)
print(numbers[:3])    # First 3 elements
print(numbers[2:])    # From element 2 to end
````

````python
# Length and membership
print(len(numbers))       # 5
print(3 in numbers)       # True
print(99 in numbers)      # False
````

````python
# What happens if you go out of bounds?
numbers[10]
````

**IndexError.** Python tells you the index is out of range. A list of 5 elements has valid indices 0–4.

## Lists are mutable

Unlike integers and strings, lists **can** be changed after creation:

````python
fruits = ["apple", "banana", "cherry"]
print(f"Before: {fruits}")
print(f"id:     {id(fruits)}")

# Change an element
fruits[1] = "blueberry"
print(f"After:  {fruits}")
print(f"id:     {id(fruits)}")   # Same id — same object, changed contents
````

The `id()` didn't change. We modified the object itself, not created a new one. This is **mutation**.

## Mutating methods

Lists have methods that modify them in place:

````python
patients = ["Alice", "Bob"]
print(f"Start: {patients}")

patients.append("Carol")       # Add to end
print(f"After append: {patients}")

patients.insert(1, "David")    # Insert at index 1
print(f"After insert: {patients}")

patients.remove("Bob")         # Remove by value
print(f"After remove: {patients}")

last = patients.pop()          # Remove and return last element
print(f"Popped: {last}")
print(f"After pop: {patients}")
````

### The `None` trap

**Mutating methods return `None`**, not the modified list. This is a very common gotcha:

````python
numbers = [3, 1, 4, 1, 5]

# WRONG — sorted_list is None!
sorted_list = numbers.sort()
print(f"sorted_list = {sorted_list}")
print(f"numbers = {numbers}")
````

`.sort()` sorted `numbers` in place and returned `None`. If you assigned the result, you lost the list.

If you want a sorted copy without modifying the original, use `sorted()` (not a method — a function):

````python
numbers = [3, 1, 4, 1, 5]
sorted_copy = sorted(numbers)
print(f"original: {numbers}")
print(f"sorted:   {sorted_copy}")
````

## The aliasing trap

This is where Week 1's reference model pays off:

````python
original = [1, 2, 3]
alias = original    # NOT a copy — same object

alias.append(4)

print(f"original: {original}")
print(f"alias:    {alias}")
print(f"Same object? {original is alias}")
````

**Both changed.** `original` and `alias` are two names for the same list. Mutating through one name is visible through the other.

This is a **very common source of bugs** in AI-generated code. The fix:

````python
original = [1, 2, 3]
safe_copy = original.copy()   # New object

safe_copy.append(4)

print(f"original:  {original}")
print(f"safe_copy: {safe_copy}")
print(f"Same object? {original is safe_copy}")
````

## Iterating over lists

````python
# Basic iteration
names = ["Alice", "Bob", "Carol"]
for name in names:
    print(f"Patient: {name}")
````

`enumerate()` gives you both the position number and the value on each loop iteration. The `i, name` syntax captures both — `i` gets the position number, `name` gets the value. This `i, name` syntax (called **unpacking**) works because `enumerate()` produces pairs of values. We'll see this pattern again with `.items()` on dictionaries in Lesson 4.2, and you'll see tuples covered in Lesson 4.3.

````python
# When you need the index too: enumerate()
for i, name in enumerate(names):
    print(f"Patient {i+1}: {name}")
````

## List comprehensions

A compact way to create a new list from an existing one:

````python
# Long way: build with a loop
numbers = [1, 2, 3, 4, 5]
squares = []
for n in numbers:
    squares.append(n ** 2)
print(squares)
````

````python
# Short way: list comprehension
squares = [n ** 2 for n in numbers]
print(squares)
````

`zip()` pairs corresponding elements from two or more iterables by position: the first element of each, then the second element of each, and so on. If the iterables have different lengths, `zip()` stops when the shortest one is exhausted. It is used when you have related data in separate sequences — for example, separate lists of weights and heights for the same patients.

````python
# With a filter
evens = [n for n in numbers if n % 2 == 0]
print(evens)
````

If you get stuck on an exercise, re-read the section above it — each exercise applies the concepts directly above. Getting stuck on new material is entirely normal; working through it carefully is how the understanding forms.

AI-generated code uses comprehensions frequently. The pattern is:

`[expression for item in iterable]` — transform each item

`[expression for item in iterable if condition]` — transform and filter

````python
# Real example: compute BMIs from data
weights = [70, 85, 62, 95, 78]
heights = [1.75, 1.80, 1.60, 1.85, 1.70]

bmis = [w / (h ** 2) for w, h in zip(weights, heights)]
for i, bmi in enumerate(bmis):
    print(f"Patient {i+1}: BMI {bmi:.1f}")
````

## What you learned

Lists are the most common data structure in Python code — you'll encounter them in virtually every program you read or review.

- Lists are ordered, zero-indexed, and **mutable**
- Mutating methods (`.append()`, `.sort()`, etc.) return `None`
- `list2 = list1` creates an alias, not a copy — use `.copy()` for independence
- List comprehensions: `[expr for item in items if condition]`
- `id()` stays the same after mutation — same object, changed contents
