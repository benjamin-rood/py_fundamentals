## Everything is an object

When you write `x = 5`, here's what Python actually does:

1. **Creates an object** — an integer containing the value `5`
2. **Stores that object in memory** at a specific address
3. **Makes the name `x` refer to that address**

The object is the real thing. The variable name is a label pointing to the object. Why does this matter? Because in the next lesson you'll see that when you write `y = x`, you're not copying the value — you're making a second label pointing to the same object. Understanding objects is what makes that behaviour make sense.

## Seeing objects with `id()`

Your computer stores all data in memory — a region of memory where Python stores every object it creates while your code runs. Every object in memory has a unique address. The `id()` function shows you that address:

```python
x = 5
print(id(x))
```

That big number is the memory address where Python stored the integer object `5`. This address will differ between runs — memory addresses are assigned at runtime and vary each time Python starts.

Let's see the address alongside the type and value:

The `f` before the opening quote and the `{braces}` let you embed variable values directly inside a string. We'll cover f-strings properly in Lesson 1.3 — for now, just read the output and notice how the values fill in.

```python
x = 5
print(f"Value: {x}")
print(f"Type:  {type(x)}")
print(f"ID:    {id(x)}")
```

Three things fully describe any Python object:
- **Value** — what it contains (`5`)
- **Type** — what kind of object it is (`int`)
- **Identity** — where it lives in memory (`id()`)

## Every type of data is an object

Integers, strings, decimals, booleans — all objects, all with addresses:

`.__name__` is an attribute that returns the type name as a plain string — `'int'` instead of the full `<class 'int'>` output.

```python
# Integer
a = 42
print(f"a = {a}, type = {type(a).__name__}, id = {id(a)}")
```

```python
# Float
b = 3.14
print(f"b = {b}, type = {type(b).__name__}, id = {id(b)}")
```

```python
# String
c = "hello"
print(f"c = {c}, type = {type(c).__name__}, id = {id(c)}")
```

Every one of these has a unique memory address. They're all real objects stored somewhere in memory.

`is` checks whether two variables point to the exact same object in memory. We'll cover `is` vs `==` properly in the next lesson.

## Small integer caching — a detail worth knowing

Small numbers like 0, 1, 2 are used constantly in almost every program. Creating a new object each time would waste time and memory, so Python creates them once at startup and reuses them. For small integers (-5 through 256), it reuses the same object instead of creating a new one each time. Watch:

```python
# Small integers: Python reuses the same object
a = 5
b = 5
print(f"id(a) = {id(a)}")
print(f"id(b) = {id(b)}")
print(f"Same object? {a is b}")
```

`a is b` is `True` — they are the same object in memory. Python did not create two separate `5`s.

Now try a larger number:

```python
# Large integers: Python creates separate objects
# Using a variable in the expression prevents Python from
# recognising these as the same value at compile time
base = 100
a = base * 10   # 1000
b = base * 10   # 1000
print(f"a = {a}, b = {b}")
print(f"id(a) = {id(a)}")
print(f"id(b) = {id(b)}")
print(f"Same object? {a is b}")
```

If you get stuck on an exercise, re-read the section above it — each exercise applies the concepts directly above. Getting stuck on new material is entirely normal; working through it carefully is how the understanding forms.

**Why `base * 10` instead of just `1000`?** When Python sees two identical number literals like `a = 1000` and `b = 1000` written in the same cell, it may optimise by reusing a single object — making `a is b` return `True` even for large numbers. Using a variable in the expression (`base * 10`) forces Python to compute the value at runtime, which reliably creates separate objects. This is a compiler-level optimisation relevant only to this demonstration.

The important takeaway: small integers (-5 to 256) are **always** cached. Larger integers **may or may not** be the same object, depending on how they were created. This is why you should use `==` to compare values and reserve `is` for identity checks.
