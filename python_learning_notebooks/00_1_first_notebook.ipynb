## What is a notebook?

This document is a notebook. It has two kinds of **cells**:

- **Text cells** (like this one): contain explanations. You just read these.
- **Code cells** (like the gray boxes below): contain Python code you can run.

To run a code cell:
1. Click on it
2. Press the **▶ play button** on the left, or press **Shift+Enter**
3. The output appears directly below the cell

## Your first code

Click on the cell below, then press Shift+Enter (or click the ▶ button).

Lines starting with `#` are **comments** — notes for humans. Python ignores them completely. Why write something Python ignores? Because code is read by people — your future self, colleagues, or an AI assistant reviewing your work. Comments explain your *intent*, which the code alone doesn't show.

```python
# Cell 1 — Click the play button or press Shift+Enter
2 + 3
```

You should see `5` appear below the cell. Here's what just happened:

1. You sent the code `2 + 3` to Python
2. Python evaluated it
3. The result (`5`) appeared below

That's the entire cycle: **write code → run it → see the result.** Everything else builds on this.

You might wonder why Python uses `**` for exponentiation instead of `^` (which you may know from maths or Excel). In Python, `^` already has a different meaning (a bitwise operation not covered in this course). So Python uses `**` to avoid the clash.

Notice: the last expression in each cell is automatically displayed below it. This is a notebook convenience — we'll see later that Python scripts behave differently.

## What happens when things go wrong?

Run this cell:

```python
# This will produce an error — that's intentional
10 / 0
```

You just saw a **ZeroDivisionError**. Python cannot divide by zero, so it reports the error.

Errors are not failures — they are Python reporting what went wrong. You'll learn to read them in Week 6. Red text with an error name means Python could not execute the instruction, and states why.

When you see an error, focus on the **last line** — that is the actual error message. The lines above it show where in the code the error occurred, which becomes relevant later.

This skill — reading what Python reports when execution fails — is one of the most practical things you'll learn. When you later review code that an AI has generated, error messages are how you determine whether the code actually works.
