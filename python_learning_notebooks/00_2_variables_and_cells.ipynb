## Variables

A **variable** is a name that refers to a value. You create one with `=`.

In Python, `=` doesn't mean "equals" like in maths. It means "make this name refer to this value." You'll sometimes hear it called **assignment**. If you want to check whether two things are equal (in the maths sense), Python uses `==` — you'll see this in Week 2.

```python
# Create a variable
x = 10
```

Nothing appeared below that cell. Assignment (`=`) doesn't display anything — it just creates the reference.

To see the value, put the variable name on its own line:

```python
# See the value
x
```

You should see `10`. The variable `x` refers to the value `10`.

Now here's something important — **cells share state.** The variable `x` you created in the cell above is still available:

```python
# Variables from earlier cells are available
x + 5
```

Python remembered that `x` refers to `10`, so `x + 5` evaluates to `15`.

### Variables can hold different kinds of data

Text must be wrapped in quotes — either `"double"` or `'single'`, both work the same. Without quotes, Python would think `John` is a variable name.

```python
# A string (text) — use quotes
name = "John"
name
```

```python
# A decimal number
height = 1.75
height
```

```python
# A True/False value
is_active = True
is_active
```

`True` and `False` must be capitalised exactly — `true` or `TRUE` won't work.

Python variable names can't have spaces, so we use underscores: `is_active` instead of `is active`. This convention is called **snake_case**.

## Displaying multiple values with `print()`

A cell only auto-displays the **last** expression. To see multiple values, use `print()`:

```python
# Without print — only the last value shows
x = 10
y = 20
x
y
```

Only `20` appeared — the last expression. Now try with `print()`:

```python
# With print — everything shows
x = 10
y = 20
print(x)
print(y)
print(x + y)
```

`print()` is a **function** — it tells Python to display a value. You'll learn much more about functions in Week 3. Use `print()` when you need to display more than one value. The parentheses `()` are how you tell Python to *run* a function — `print` is the function's name, and `()` means "execute it now with these inputs." You'll see this pattern everywhere: `type()`, `len()`, `id()`.

## What type is this? — `type()`

Every value in Python has a **type** — what kind of data it is. The `type()` function tells you:

```python
type(5)
```

```python
type(3.14)
```

```python
type("hello")
```

```python
type(True)
```

The four basic types you'll see most often:

| Type | Name | Examples |
|------|------|----------|
| `int` | Integer | `5`, `-3`, `0`, `1000` |
| `float` | Decimal number | `3.14`, `-0.5`, `2.0` |
| `str` | String (text) | `"hello"`, `"42"`, `""` |
| `bool` | Boolean (True/False) | `True`, `False` |

Notice: `"42"` is a **string** (text), not an integer. The quotes make it text.

```python
# These are different types!
print(type(42))
print(type("42"))
```

Why does this matter? Because the type determines which operations are valid on a value. `42 + 1` gives you `43`, but `"42" + 1` raises a TypeError — Python cannot add an integer to a string. When something goes wrong, checking the type with `type()` is often the first step in diagnosing why.

## ⚠️ THE CRITICAL RULE: Execution Order

This is the most important thing to understand about notebooks.

**Cells run in the order YOU click them, not the order they appear on screen.**

This can cause confusion. Let's see it happen.

A regular Python program (which you'll use from Week 5) runs from top to bottom, once, in order. A notebook is different — it is an interactive environment where any cell can be run at any time, in any order. That flexibility is powerful, but it means the order on screen and the order of execution can get out of sync.

**Step 1:** Run this cell:

```python
message = "hello"
```

**Step 2:** Run this cell:

```python
print(message)
```

### The fix: Restart and run all

When things seem confusing or out of sync:

In the menu bar at the top of the page, click **Runtime**, then select **Restart and run all**.

This runs every cell from top to bottom, starting completely fresh. It's the single best debugging tool you have in a notebook.

**Develop this habit. Use it often.** Professional programmers who use notebooks do this constantly.
