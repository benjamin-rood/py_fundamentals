## The classification

Every Python type is either **mutable** (can be changed) or **immutable** (cannot be changed):

| Immutable (can't change) | Mutable (can change) |
|--------------------------|---------------------|
| `int` | `list` |
| `float` | `dict` |
| `str` | `set` |
| `bool` | |
| `tuple` | |

> **Note on `set`:** Sets are unordered collections with no duplicates. We won't use them in this course, but you may see them in AI-generated code.

## Why it matters: the aliasing rule

This is where the reference model from Week 1 pays off. Remember: variables don't *contain* values — they *refer to* objects in memory. When two variables refer to the same object, what happens next depends entirely on whether that object is mutable or immutable.

**Aliasing is safe with immutable objects.** Two names pointing to the same integer or string can never cause bugs — you can't change the object, only rebind the name.

**Aliasing is dangerous with mutable objects.** Two names pointing to the same list or dict means changing it through one name affects the other.

````python
# Safe: immutable aliasing
a = 10
b = a
a = a + 1
print(f"a = {a}, b = {b}")  # b is unaffected
````

````python
# Dangerous: mutable aliasing
a = [1, 2, 3]
b = a
a.append(4)
print(f"a = {a}")
print(f"b = {b}")  # b changed too!
````

## Why it matters: functions

This is the same principle applied to function parameters:

````python
def process_immutable(x):
    x = x + 1       # Reassignment — new object
    return x

value = 10
result = process_immutable(value)
print(f"value = {value}")   # Still 10 — unaffected
print(f"result = {result}")  # 11
````

````python
def process_mutable(items):
    items.append("added")  # Mutation — same object

my_list = ["original"]
process_mutable(my_list)
print(f"my_list = {my_list}")  # Changed! ["original", "added"]
````

### How to protect against accidental mutation

If a function should NOT modify the caller's data, copy first:

````python
def safe_process(items):
    working_copy = items.copy()
    working_copy.append("added")
    return working_copy

my_list = ["original"]
result = safe_process(my_list)
print(f"my_list = {my_list}")  # Unaffected
print(f"result = {result}")   # ["original", "added"]
````

## Tuples — immutable sequences (brief)

A **tuple** is an immutable ordered sequence. Tuples use parentheses `()` while lists use square brackets `[]`. These parentheses look like mathematical grouping, but here they create a data structure:

````python
point = (3, 7)
print(point[0])   # 3 — access works
print(len(point)) # 2
````

````python
# Can't modify
point[0] = 5
````

Why would you choose a tuple over a list? A tuple signals *intent*: "this data should not change." When you see a tuple in code, you know it's meant to be a fixed collection — like a coordinate pair or a date. Lists, by contrast, signal "this collection may grow, shrink, or be reordered." Tuples appear in AI-generated code as return values (functions returning multiple things) and as dict keys (since dict keys must be immutable).

## What you learned

- Immutable: `int`, `float`, `str`, `bool`, `tuple` — can't be changed, only replaced
- Mutable: `list`, `dict`, `set` — can be changed in place
- Aliasing (two names, same object) is safe with immutable, dangerous with mutable
- In functions: reassignment is always local; mutation affects the caller
- Protect against accidental mutation with `.copy()`
