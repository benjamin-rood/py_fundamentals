## `for` loops with `range()`

`range(n)` produces numbers from 0 to n-1. A `for` loop runs the indented body once for each number:

```python
for i in range(5):
    print(i)
```

Starting from 0 is a convention throughout programming — list indices, ranges, and more all start from 0.

`range(5)` gave us 0, 1, 2, 3, 4 — five numbers, starting from 0.

Why does `range(5)` stop at 4 instead of including 5? Because the number you pass *is* the count: `range(5)` gives exactly 5 items. And `range(1, 6)` gives exactly `6 - 1 = 5` items. The rule is always: the stop value is excluded, so the *length* of the range equals `stop - start`. This same convention applies to list slicing — once you learn it, it works the same everywhere in Python.

You can also specify start and stop:

```python
# range(start, stop) — stop is NOT included
for i in range(1, 6):
    print(i)
```

```python
# range(start, stop, step)
for i in range(0, 20, 5):
    print(i)
```

### Doing something useful with a loop

```python
# Sum the numbers 1 through 100
total = 0
for i in range(1, 101):
    total = total + i
print(f"Sum of 1–100: {total}")
```

```python
# Build up a string
result = ""
for i in range(1, 6):
    result = result + f"Patient {i}, "
print(result)
```

## `for` loops over lists (preview)

You'll learn lists properly in Week 4, but the basic iteration is straightforward:

```python
names = ["Alice", "Bob", "Carol"]

for name in names:
    print(f"Checking patient: {name}")
```

`for name in names:` means: for each element in the list, call it `name` and run the body.

```python
# Combining loops with conditionals
temperatures = [36.5, 38.2, 37.0, 39.1, 36.8]

for temp in temperatures:
    if temp > 37.5:
        print(f"{temp}°C — FEVER")
    else:
        print(f"{temp}°C — normal")
```

## `while` loops

A `while` loop repeats as long as a condition is `True`:

```python
# Countdown
count = 5
while count > 0:
    print(count)
    count = count - 1
print("Done!")
```

⚠️ **Critical:** The condition must eventually become `False`, or the loop runs forever. `count = count - 1` is what makes `count > 0` eventually `False`.

## `break` and `continue`

- `break` — exit the loop immediately
- `continue` — skip to the next iteration

```python
# break — stop when you find what you're looking for
numbers = [3, 7, 2, 8, 1, 9, 4]

for n in numbers:
    if n > 8:
        print(f"Found a number > 8: {n}")
        break
    print(f"Checked {n}, not > 8")
```

`%` gives the remainder when dividing. For example, `7 % 2` is `1` (because 7 divides into 2 three times with 1 left over). If the remainder is `0`, the number divides evenly — so `i % 2 == 0` checks whether `i` is even.

```python
# continue — skip certain items
for i in range(10):
    if i % 2 == 0:   # % is modulo — remainder after division
        continue     # Skip even numbers
    print(i)         # Only odd numbers print
```

### Mini-project: Number guessing game

This combines `while`, `if/elif/else`, `input()`, and `break`.

`import random` loads one of Python's built-in **modules** — a module is a file of pre-written code organised by topic. Python comes with many modules (for maths, dates, file handling, etc.), but it doesn't load them all automatically — that would slow every program down. Instead, you load only what you need with `import`. We'll cover this properly in Week 5. For now, just know that `import random` makes `random.randint()` available.

`while True:` creates a loop that would run forever on its own — but we use `break` inside to exit when the guess is correct. This "loop forever, break when done" pattern is common in Python.

```python
age_text = input("Enter age: ")
age = int(age_text)    # Convert string to integer
print(f"Age: {age}, type: {type(age)}")
```

Why do we need `int(age_text)`? Because `input()` *always* returns a string — even if the user types `34`, Python receives the text `"34"`, not the number `34`. Python doesn't assume what text from the outside world means. You have to explicitly convert it with `int()` (for whole numbers) or `float()` (for decimals). If the user types something that isn't a number, the conversion will fail with a `ValueError` — which is Python protecting you from bad data.

### Mini-project: Number guessing game

This combines `while`, `if/elif/else`, `input()`, and `break`:

If you are stuck on an exercise for more than a few minutes, re-read the section above it — each exercise applies the concepts in that section.

```python
import random

secret = random.randint(1, 20)
attempts = 0

print("I'm thinking of a number between 1 and 20.")

while True:
    guess_text = input("Your guess: ")
    guess = int(guess_text)
    attempts = attempts + 1

    if guess < secret:
        print("Too low!")
    elif guess > secret:
        print("Too high!")
    else:
        print(f"Correct! It was {secret}. You got it in {attempts} attempts.")
        break
```
