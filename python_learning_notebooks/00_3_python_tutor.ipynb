# Lesson 0.3: Python Tutor

**Duration:** ~30 minutes

Python Tutor is a website that shows you what Python is doing **step by step**. Your notebook shows you *what* happened (the output). Python Tutor shows you *how* and *why*.

## What Python Tutor shows you

- **Objects in memory** — the actual data Python created
- **Variables pointing to objects** — which name refers to which data
- **Execution order** — which line runs next

When you step through code, pay attention to the arrows between variable names (on the left) and objects (on the right). Watch which arrow moves when a line executes — that's the key to understanding what each line actually *does*.

This will become invaluable in Week 1, when we look at how variables and objects really work.

## Try it now

1. Open a new browser tab
2. Go to [pythontutor.com](https://pythontutor.com)
3. Paste this code into the editor. You'll see a text box where you can type or paste code:

```python
x = 5
y = x
x = 10
print(y)
```

4. Click **"Visualize Execution"**
5. Click the **"Next"** button to step through one line at a time
6. Watch the variables and objects change as each line executes

## What did you see?

When you stepped through:

1. `x = 5` — Python creates the value `5` and `x` points to it
2. `y = x` — `y` now points to the **same** value `5` (not a copy)
3. `x = 10` — `x` now points to a **new** value `10`. But `y` still points to `5`.
4. `print(y)` — outputs `5`

This is a preview of Week 1's big idea: **variables point to objects.** Reassigning `x` doesn't affect `y` because `y` has its own reference to `5`.

## When to use Python Tutor

Later in this course, you'll be reading code that an AI has written. Python Tutor is how you check whether that code does what you think it does — by watching it execute, step by step.

Use Python Tutor when:
- You're confused about **why** code behaves a certain way
- You want to see **what's happening in memory**, not just the output
- You need to trace through a loop or function call step by step

**Workflow for this course:**
1. Try something in a notebook cell
2. If confused about **why** → paste into Python Tutor
3. Step through and watch
4. Return to notebook to experiment more

Links to Python Tutor will appear throughout the course where visualization is especially helpful.
