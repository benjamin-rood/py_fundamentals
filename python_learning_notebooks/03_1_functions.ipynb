# Lesson 3.1: Defining and Calling Functions

**Duration:** ~1.5 hours

Functions let you define a reusable block of code with a name. You've been *calling* functions (`print()`, `type()`, `id()`, `len()`). Now you'll learn to *write* them.

Why do functions exist? Imagine you need to calculate BMI for 50 patients. Without functions, you'd copy and paste the same formula 50 times — and if the formula changed, you'd need to update all 50 copies. A function lets you write the logic once, give it a name, and reuse it wherever you need it. Functions also make code easier to read: `calculate_bmi(weight, height)` tells you what's happening without you needing to read the formula.

## Defining a function

````python
def greet(name):
    """Print a greeting for the given name."""
    print(f"Hello, {name}!")
````

That cell defined a function but didn't run it. Nothing happened yet.

- `def` (short for "define") starts the definition
- `greet` is the function's name
- `name` is a **parameter** — a placeholder for the value you'll pass in
- The indented lines are the **body** — the code that runs when you call it
- The triple-quoted string is a **docstring** — what the function does

## Calling a function

````python
greet("Alice")
greet("Bob")
greet("Carol")
````

When you call `greet("Alice")`, Python:
1. Creates a local variable `name` referring to the string `"Alice"`
2. Runs the body with that `name`
3. Returns control to the caller when the body finishes

`"Alice"` is the **argument** — the actual value you passed in. `name` is the **parameter** — the variable that receives it.

A memory aid: **P**arameter = **P**laceholder (in the definition). **A**rgument = **A**ctual value (in the call).

## Returning values

Functions that compute something should **return** the result, not just print it:

````python
def calculate_bmi(weight_kg, height_m):
    """Calculate Body Mass Index."""
    bmi = weight_kg / (height_m ** 2)
    return bmi
````

````python
result = calculate_bmi(70, 1.75)
print(f"BMI: {result:.1f}")
````

````python
# You can use the return value directly
print(f"BMI: {calculate_bmi(85, 1.80):.1f}")
````

Notice that `calculate_bmi(85, 1.80)` appears directly inside the f-string. A function call is an expression — it produces a value, and that value can be used anywhere you'd use a number, string, or variable: inside `print()`, in a calculation, as part of an f-string, or as an argument to another function.

### `return` vs `print` — why it matters

`print()` shows a value on screen. `return` sends a value back to the caller so it can be used.

Watch the difference:

````python
# This function PRINTS — you can see it but can't use the result
def bmi_print(weight, height):
    bmi = weight / (height ** 2)
    print(f"BMI: {bmi:.1f}")

result1 = bmi_print(70, 1.75)
print(f"result1 = {result1}")   # None!
````

````python
# This function RETURNS — you can use the result
def bmi_return(weight, height):
    bmi = weight / (height ** 2)
    return bmi

result2 = bmi_return(70, 1.75)
print(f"result2 = {result2:.1f}")   # 22.9
````

The print version returned `None` — the special Python value meaning "nothing." A function without `return` (or with just `return` by itself) always returns `None`.

This is the same pattern you saw with strings: `name.upper()` returns a new value, but if you don't capture it, it's lost. Here, `bmi_print()` executes (printing to screen) but returns no value — so Python fills in `None` as the return value. Whenever you see `None` where you expected a real value, check whether a `return` call is missing from the function body:

```python
def double(x):
    result = x * 2   # computed but never returned

print(double(5))     # None
```

## Docstrings

The triple-quoted string right after `def` is the **docstring**. It documents what the function does:

````python
def categorize_bmi(bmi):
    """Categorize a BMI value into a risk category.

    Args:
        bmi: Body Mass Index as a float

    Returns:
        String category: 'underweight', 'normal', 'overweight', or 'obese'
    """
    if bmi < 18.5:
        return "underweight"
    elif bmi < 25:
        return "normal"
    elif bmi < 30:
        return "overweight"
    else:
        return "obese"
````

````python
# You can read the docstring with help()
help(categorize_bmi)
````

Docstrings aren't required, but they're valuable — especially when reviewing AI-generated code. If a function has a good docstring, you can quickly understand what the function is supposed to do before reading the implementation.

This format (listing Args and Returns) is a common convention called **Google-style docstrings**. You'll see it frequently in AI-generated code.

## Multiple parameters and return values

````python
def patient_summary(name, weight_kg, height_m):
    """Generate a summary string for a patient."""
    bmi = weight_kg / (height_m ** 2)
    category = categorize_bmi(bmi)
    return f"Patient {name}: BMI {bmi:.1f} ({category})"

print(patient_summary("Alice", 70, 1.75))
print(patient_summary("Bob", 95, 1.82))
````

Notice: `patient_summary` calls `categorize_bmi` — functions can call other functions. This is how real programs are built: small functions that each do one thing, composed together.
