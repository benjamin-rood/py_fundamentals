## Boolean values

A **boolean** is a value that's either `True` or `False`. That's it â€” only two possible values.

````python
print(type(True))
print(type(False))
````

## Comparison operators

Comparisons produce boolean values:

````python
print(5 == 5)     # Equal to
print(5 != 3)     # Not equal to
print(5 > 3)      # Greater than
print(5 < 3)      # Less than
print(5 >= 5)     # Greater than or equal
print(5 <= 3)     # Less than or equal
````

### `is` vs `==` (from Week 1)

Quick reminder:
- `==` checks if two values are **equal**
- `is` checks if two variables point to the **same object**

For checking `None`, always use `is`:

```python
if result is None:      # âś… Correct
if result == None:      # âťŚ Works but wrong habit
```

`None` is Python's special value meaning "nothing" â€” you'll encounter it properly in Week 3 when we cover functions. Why `is` and not `==`? Because there is exactly one `None` object in all of Python â€” it's a **singleton**. Checking `is None` asks "is this the one-and-only None object?" which is both faster and more precise than comparing values with `==`.

## Making decisions: `if` / `elif` / `else`

The `if` statement runs code only when a condition is `True`:

````python
temperature = 38.5  # Celsius

if temperature > 37.5:
    print("Fever detected")
    print("Alert the physician")
````

**Indentation matters.** The lines indented under `if` are the **body** â€” they only run if the condition is `True`. Use 4 spaces (Colab does this automatically when you press Tab).

Notice the colon `:` at the end of the `if` line â€” it's required. Python uses it to mark the start of an indented block. Most other programming languages use curly braces `{}` to mark which lines belong to an `if` block. Python uses indentation instead â€” the idea is that the code's visual structure *is* its logical structure. You never have to guess which lines are inside the `if`, because you can see it directly.

````python
# if / else â€” two branches
temperature = 36.8

if temperature > 37.5:
    print("Fever detected")
else:
    print("Temperature normal")
````

````python
# if / elif / else â€” multiple branches
bmi = 27.3

if bmi < 18.5:
    category = "underweight"
elif bmi < 25:
    category = "normal"
elif bmi < 30:
    category = "overweight"
else:
    category = "obese"

print(f"BMI {bmi} â†’ {category}")
````

Python checks conditions **top to bottom** and takes the **first** branch that's `True`. Once a branch executes, the rest are skipped.

Try changing `bmi` to `18.0`, then `24.5`, then `32.1`. Run the cell each time.

## Combining conditions: `and`, `or`, `not`

````python
age = 65
has_diabetes = True

# Both must be True
if age > 60 and has_diabetes:
    print("High risk patient")

# Either can be True
if age > 60 or has_diabetes:
    print("Needs monitoring")

# Flip True to False
if not has_diabetes:
    print("No diabetes")
else:
    print("Has diabetes")
````

## Truthy and Falsy (brief)

Python treats some values as `False` in boolean contexts:

Why does Python do this? Writing `if name != "":` or `if len(items) != 0:` to check for empty data is verbose. Python defines a rule: zero, empty sequences, and `None` are **falsy** (evaluate to `False` in a boolean context), and everything else is **truthy** (evaluates to `True`).

- `0`, `0.0` (zero)
- `""` (empty string)
- `[]` (empty list)
- `None`
- `False`

Everything else is treated as `True`. This means you can write:

This matters because AI-generated code often writes `if name:` instead of `if name != "":` â€” they are equivalent, and now you understand the mechanism.

````python
name = "Alice"

if name:    # True because name is not empty
    print(f"Hello, {name}")
else:
    print("No name provided")

# Try changing name to "" (empty string) and run again
````
