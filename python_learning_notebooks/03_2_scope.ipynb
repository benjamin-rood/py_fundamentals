When you call a function, Python creates a temporary workspace for it. This lesson covers what lives in that workspace and what happens when the function ends.

Why does Python create a separate workspace for each function? Without it, every function would share the same pool of variable names. If two functions both used a variable called 'result', they'd overwrite each other's data. Scope keeps functions independent — each one gets its own workspace where its variables can't accidentally interfere with anything else.

## Local variables disappear when the function returns

````python
def calculate(x, y):
    result = x + y    # 'result' is local to this function
    return result

answer = calculate(10, 20)
print(f"answer = {answer}")
````

````python
# Try to access the local variable — it doesn't exist out here
print(result)
````

**NameError: name 'result' is not defined.** The variable `result` was local — it only existed inside `calculate()` while it was running. When the function returned, that workspace was destroyed.

This is called **scope.** Variables created inside a function are in the function's scope, not the global scope.

Notice the structure of the error: 'NameError' tells you the *category* of problem (Python can't find a name), and 'name \'result\' is not defined' tells you *which* name and *what's wrong*. Most Python errors follow this pattern: the error type gives you the category, and the message gives you the specifics. Learning to read this structure is one of the most useful debugging skills you'll develop.

## Stack frames

Each function call creates a **frame** — a temporary workspace:

1. You call `calculate(10, 20)`
2. Python creates a frame with: `x → 10`, `y → 20`, then `result → 30`
3. The function returns `30`
4. The frame is destroyed — `x`, `y`, `result` are gone
5. Back in the caller, `answer → 30`

You may hear this workspace called a "stack frame" or just "frame." The name isn't important — think of it as a temporary whiteboard that the function uses while it runs. When the function finishes, the whiteboard is erased and all its local variables are gone.

## Variable shadowing

A local variable can have the same name as a global one. The local one takes priority inside the function:

````python
x = 100   # global

def show_x():
    x = 999   # local — shadows the global x
    print(f"Inside function: x = {x}")

show_x()
print(f"Outside function: x = {x}")
````

The function's `x = 999` is a separate variable from the global `x = 100`. They just happen to share a name. When the function ends, its local `x` disappears and the global `x` is untouched.

This is worth knowing because if you accidentally reuse a global variable's name inside a function, your function will use its own local version — which can produce unexpected behavior that is difficult to trace.

You might wonder: can a function *deliberately* modify a global variable? Python has a 'global' keyword that allows this — but it's rarely used and generally considered bad practice. If a function needs to affect the outside world, it's cleaner to 'return' a value and let the caller decide what to do with it. You may see 'global' in AI-generated code occasionally; now you know what it does.

## Parameter passing: references, not copies

When you pass an argument to a function, the parameter receives a **reference** to the same object — not a copy. We can verify this with `id()`:

````python
def examine(value):
    print(f"  Inside function: value = {value}, id = {id(value)}")

x = 42
print(f"Before call: x = {x}, id = {id(x)}")
examine(x)
print(f"After call:  x = {x}, id = {id(x)}")
````

Same `id()` inside and outside — the parameter `value` and the argument `x` refer to the same object.

But what happens if the function **reassigns** the parameter?

````python
def try_to_change(value):
    print(f"  Before reassignment: value = {value}, id = {id(value)}")
    value = value + 1   # Reassignment — new object
    print(f"  After reassignment:  value = {value}, id = {id(value)}")

x = 42
print(f"Before call: x = {x}, id = {id(x)}")
try_to_change(x)
print(f"After call:  x = {x}, id = {id(x)}")
````

Key observations:

- Before reassignment: `value` and `x` have the **same** `id()` — same object
- After `value = value + 1`: `value` has a **new** `id()` — it now refers to a different object (`43`)
- After the call: `x` is still `42` — the reassignment inside the function didn't affect the caller

**Reassignment inside a function never changes the caller's variable.** This is because integers are immutable — `value + 1` creates a new object and rebinds the local name `value` to it.

## What about mutable objects?

With mutable objects (like lists), the situation is different. But we haven't covered lists in depth yet — that's Week 4. Here's a quick preview:

````python
def add_item(my_list):
    my_list.append("new item")   # Modifies the object itself

data = ["original"]
print(f"Before: {data}")
add_item(data)
print(f"After:  {data}")
````

The function **modified** the list — and the caller sees the change! That's because:
- `my_list` refers to the **same** list object as `data` (same `id()`)
- `.append()` changes the list **in place** (mutation)
- There's no reassignment — `my_list` still refers to the same object

You'll understand this fully in Week 4 when we cover mutability. For now, the rule is:

- **Reassignment** inside a function → caller is **NOT** affected
- **Mutation** inside a function → caller **IS** affected
