Dictionaries store data as key-value pairs. Lists of dictionaries are the most common data shape in real applications — APIs, databases, and AI-generated code all use this pattern. Why not just use a list for everything? A list is good when you have an ordered sequence of similar items (like a list of temperatures). But when each piece of data has a *name* — a patient's age, weight, blood type — you need a structure where you can ask for data by name rather than by position number. That's what a dictionary is: a collection of named values.

## Creating and accessing dictionaries

```python
patient = {
    "name": "Alice",
    "age": 34,
    "weight": 70,
    "height": 1.75
}

print(patient)
print(type(patient))
```

```python
# Access by key
print(patient["name"])
print(patient["age"])
```

```python
# What happens with a missing key?
print(patient["blood_type"])
```

**KeyError.** The key doesn't exist. Use `.get()` for safe access:

```python
# .get() returns None (or a default) instead of crashing
print(patient.get("blood_type"))          # None
print(patient.get("blood_type", "unknown"))  # "unknown"
```

## Dictionaries are mutable

```python
patient["blood_type"] = "A+"    # Add a new key
patient["age"] = 35              # Update an existing key
print(patient)
```

```python
del patient["blood_type"]        # Remove a key
print(patient)
```

`del` removes something — here it deletes a key (and its value) from the dictionary.

## Iterating over dictionaries

```python
patient = {"name": "Alice", "age": 34, "weight": 70}

# Keys
for key in patient:
    print(key)

print()
```

```python
# Values
for value in patient.values():
    print(value)

print()
```

```python
# Both: .items() gives you (key, value) pairs
for key, value in patient.items():
    print(f"{key}: {value}")
```

The `key, value` in `for key, value in patient.items():` uses the same unpacking syntax you saw with `enumerate()` in Lesson 4.1. Each call to `.items()` produces a pair (key, value), and Python splits that pair into two separate variables for you. Tuples are covered briefly in Lesson 4.3.

## Lists of dictionaries — the "records" pattern

This is the most important data pattern in this course. Almost every real dataset looks like this:

```python
patients = [
    {"name": "Alice", "weight": 70, "height": 1.75, "risk": 1},
    {"name": "Bob",   "weight": 95, "height": 1.82, "risk": 3},
    {"name": "Carol", "weight": 55, "height": 1.60, "risk": 2},
    {"name": "David", "weight": 110,"height": 1.78, "risk": 4},
]

# Access: list index, then dict key
print(patients[0]["name"])     # Alice
print(patients[2]["weight"])   # 55
```

Notice we use single quotes `'name'` inside the f-string because the f-string itself uses double quotes. You can't use the same quote style for both.

```python
# Process all patients
for patient in patients:
    bmi = patient["weight"] / (patient["height"] ** 2)
    print(f"{patient['name']}: BMI {bmi:.1f}, Risk {patient['risk']}")
```

```python
# Filter: find high-risk patients
high_risk = [p for p in patients if p["risk"] >= 3]

print("High risk patients:")
for p in high_risk:
    print(f"  {p['name']} (risk {p['risk']})")
```

This pattern — iterating over a list of dicts, accessing keys, filtering — is what you'll see in almost every AI-generated data processing function.

## Checking for keys

```python
patient = {"name": "Alice", "weight": 70}

print("name" in patient)      # True — checks keys, not values
print("height" in patient)    # False
```

If you get stuck on an exercise, re-read the section above it — each exercise applies the concepts directly above. Getting stuck on new material is entirely normal; working through it carefully is how the understanding forms.

```python
# Safe processing with key checking
for p in patients:
    if "weight" in p and "height" in p:
        bmi = p["weight"] / (p["height"] ** 2)
        print(f"{p['name']}: BMI {bmi:.1f}")
    else:
        print(f"{p['name']}: missing data")
```
