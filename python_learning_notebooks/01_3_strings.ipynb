## f-strings — formatting text with data

An **f-string** lets you embed variables and expressions directly inside a string:

````python
name = "Alice"
age = 34
print(f"Patient: {name}, Age: {age}")
````

The `f` before the opening quote makes it an f-string. Anything inside `{}` is evaluated as Python code:

````python
weight = 70
height = 1.75
bmi = weight / (height ** 2)
print(f"BMI: {bmi}")
````

This produces more decimal places than is typically needed. You can control the formatting:

````python
# Format to 1 decimal place
print(f"BMI: {bmi:.1f}")

# Format to 2 decimal places
print(f"BMI: {bmi:.2f}")
````

Inside an f-string's `{}`, everything after the `:` is a **format specifier** — an instruction for how to display the value. In `:.1f`: the `:` separates the value from the formatting instruction, `.1` means "1 digit after the decimal point", and `f` means "display as a fixed-point (decimal) number." So `:.2f` means 2 decimal places, `:.0f` means no decimal places (round to a whole number).

You can also do math inside the braces:

````python
print(f"BMI: {weight / (height ** 2):.1f}")
````

## String methods

Strings have **methods** — built-in operations you call with a dot:

````python
text = "Hello, World"

print(text.upper())        # All uppercase
print(text.lower())        # All lowercase
print(text.replace("World", "Python"))  # Replace text
print(text.startswith("Hello"))    # True/False check
print(text.split(", "))     # Split into a list
````

## Strings are immutable — methods return NEW strings

Remember from Lesson 1.2: strings can't be modified. So what do these methods do? They create and return a **new** string:

````python
original = "hello"
upper = original.upper()

print(f"original: {original}, id = {id(original)}")
print(f"upper:    {upper}, id = {id(upper)}")
print()
print(f"original is unchanged: {original}")
````

`.upper()` didn't change `original` — it returned a completely new string. The `id()`s are different because they're different objects.

This follows a Python design convention: methods on immutable types (like strings) always return a new object, because they can't modify the original. Later, with lists, you'll see the mirror image — methods that *do* modify the object (like `.append()`) return `None` instead, as a signal that the change happened in place. Once understood, this convention makes both patterns predictable.

This is a common mistake: calling a method and forgetting to save the result:

````python
name = "alice"
name.upper()       # Creates "ALICE" but throws it away!
print(name)        # Still "alice"

# Fix: save the result
name = name.upper()  # Reassign to the new string
print(name)        # Now "ALICE"
````

## You can't modify a string by index

You can read individual characters, but you can't change them:

You can access individual characters in a string by their position using square brackets. Positions start counting at 0, not 1 — so `text[0]` is the first character, `text[1]` is the second, and so on.

````python
text = "hello"
print(text[0])     # 'h' — reading works
````

````python
# This will cause an error:
text = "hello"
text[0] = "H"
````

`text[1:]` means "everything from position 1 to the end." Reading it piece by piece: `1` is the starting position, `:` means "through to", and leaving the right side empty means "the end." If you wrote `text[1:3]`, you'd get positions 1 and 2 — the stop position is *excluded*, just like `range()`. This design is consistent throughout Python: ranges and slices always include the start but exclude the stop.

**TypeError: 'str' object does not support item assignment.** That's Python telling you: strings are immutable.

To "change" a string, create a new one:

````python
text = "hello"
new_text = "H" + text[1:]   # Slice from index 1 onward
print(new_text)
````

## Useful string operations

A few more tools you'll use often:

````python
# len() — how many characters?
print(len("hello"))     # 5
print(len(""))          # 0 (empty string)
````

````python
# strip() — remove whitespace from edges
messy = "   hello   "
print(messy.strip())    # "hello"
````

The syntax places the separator before `.join()`: `" ".join(words)` means "use the space character to join the items in `words`." The separator goes before `.join()`, and the list goes inside the parentheses.

If you get stuck on an exercise, re-read the section above it — each exercise applies the concepts directly above. Getting stuck on new material is entirely normal; working through it carefully is how the understanding forms.

````python
# in — check if a substring exists
print("ell" in "hello")     # True
print("xyz" in "hello")     # False
````

````python
# join() — combine a list of strings
words = ["Python", "is", "fun"]
print(" ".join(words))       # "Python is fun"
print(", ".join(words))      # "Python, is, fun"
````
